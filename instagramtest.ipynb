```python
# 1.Drop completely empty columns
cols_to_drop = ['discovery_input','has_handshake','location']
df=df.drop(columns=cols_to_drop)

# 2.Impute 'video_play_count' - set NaN to 0
df['video_play_count'] = df['video_play_count'].fillna(0).astype(int)

# 3.For hashtags: drop rows with missing hashtags **only for group A later**
# for now we keep them and mark missing
df['hashtags_missing'] = df['hashtags'].isna()

# 4.fill low-missing text fields
df['latest_comments']=df['latest_comments'].fillna('')
df['description'] = df['description'].fillna('')


```

```python
import numpy as np
from ast import literal_eval
```

```python
### parse 'hashtags' column from string to real Python List
df['hashtags'] = df['hashtags'].apply(lambda x: literal_eval(x) if isinstance(x,str) else [])

df[['likes','num_comments','followers']] = df[['likes','num_comments','followers']].astype(int)
```

### STEP 1 EXPLANATION
We load the CSV, remove rows with missing likes/comments/hashtags/followers because they break engagement calculations `literal_eval` safely turns the hashtag strings like "['cat','fun']"
into real Python lists we can search.We cast counts to int so math operations work correctly.

# RANDOMLY ASSIGN EACH ROW TO A/B GROUPS

```python
np.random.seed(42)
df['ab_group'] = np.random.choice(['A','B'],size=len(df),p=[0.5,0.5]) # A = hashtag‑match, B = influencer
```

```python
df
```

### STEP 2 EXPLANATION
Each post impression is randomly labelled ‘A’ or ‘B’.
A 50/50 split keeps the experiment balanced and avoids bias.
We seed NumPy’s RNG so results can be reproduced exactly.

# BUILD GROUP‑SPECIFIC FILTERS / RANKING LOGIC

```python
# Define "popular influencer threshold (top 20% followers)"
follower_threshold = df['followers'].quantile(0.8)

#Create boolean masks for each strategy
df['is_influencer_post'] = ((df['is_verified'] == True) | (df['followers']>=follower_threshold))

#simulate a list of GLOBAL top-interest hashtags
top_hashtags_global = (df.explode('hashtags').value_counts('hashtags').head(20).index.tolist())

# label rows that match those "interest" hashtags

df['matches_interest_hashtag'] = df['hashtags'].apply(lambda tag_list: any(tag in top_hashtags_global for tag in tag_list))
```

STEP 3 EXPLANATION

- We define influencer posts as either verified accounts OR accounts
-  in the top 20 % by follower count.
- For the hashtag strategy we simulate “user interests” using the 20
- most frequent hashtags in the whole dataset (you can customise this
- per user if you have interaction history).
- Two boolean columns now tell us whether each post fits strategy A or B.

# CALCULATE ENGAGEMENT METRICS

```python
# Aggregate core engagement metrics
df['engagement'] = df['likes'] + df['num_comments']

# Normalise by follower base -> engagement rate
df['engagement_rate'] = df['engagement'] / df['followers'].replace(0,np.nan)

# keep only posts that follow their assigned strategy
df_A = df[(df['ab_group'] == 'A') & (df['matches_interest_hashtag'])]
df_B = df[(df['ab_group'] == 'B') & (df['is_influencer_post'])]
```

### STEP 4 EXPLANATION
- `engagement` is likes + comments — a simple, interpretable metric.
- We divide by the poster’s follower count to avoid bias toward huge pages.
- We filter rows so each group is evaluated only on posts that actually match its own recommendation rule.

# STATISTICAL TESTING & VISUALISATION

```python
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# t-test (does not assume equal variance)
t_stat,p_val = ttest_ind(df_A['engagement_rate'].dropna(),
                         df_B['engagement_rate'].dropna(),equal_var=False)
print(f"T-statistics: {t_stat:.4f}, P-value: {p_val:.4f}")

# Visual comparision
sns.boxplot(data=pd.concat([df_A,df_B]),x='ab_group',y='engagement_rate',order=['A','B'])
plt.title('Engagement Rate by A/B Group')
plt.ylabel('Engagement Rate (likes+comments per follower)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------------------------
# Assumes you still have df_A and df_B from Step 4
# If not, run the Step 4 filters first.
# -------------------------------------------------
A = df_A['engagement_rate'].dropna()
B = df_B['engagement_rate'].dropna()

# 1️⃣  Side‑by‑side histograms  --------------------
plt.figure(figsize=(6,4))
plt.hist(A, bins=30, alpha=0.5, label='Group A')
plt.hist(B, bins=30, alpha=0.5, label='Group B')
plt.xlabel('Engagement rate')
plt.ylabel('Frequency')
plt.title('Histogram of engagement rates')
plt.legend()
plt.tight_layout()

# 2️⃣  Violin plots (with sample means)  -----------
plt.figure(figsize=(4,4))
plt.violinplot([A, B], showmeans=True)
plt.xticks([1, 2], ['Group A', 'Group B'])
plt.ylabel('Engagement rate')
plt.title('Violin plot of engagement rates')
plt.tight_layout()

# 3️⃣  Mean bar chart with 95 % CIs  ---------------
mean_A, mean_B = A.mean(), B.mean()
ci_A = 1.96 * A.std(ddof=1) / np.sqrt(len(A))   # 95 % CI
ci_B = 1.96 * B.std(ddof=1) / np.sqrt(len(B))

plt.figure(figsize=(4,4))
plt.bar([0, 1], [mean_A, mean_B],
        yerr=[ci_A, ci_B],
        capsize=5,
        tick_label=['Group A', 'Group B'])
plt.ylabel('Engagement rate')
plt.title('Mean engagement rate ± 95 % CI')
plt.tight_layout()

plt.show()
```
